Linear & Log Transformations using the Auto MPG Dataset

Linear Transformations - Made up of main operations(Scaling, Shifting, Standardizing, Centering)
Scaling - Multiplying/Dividing by a constant
Shifting - Add/subtracting a constant
Centering - Subtracting the mean
Standardizing - Subtracting the mean and dividing by the standard deviation

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('auto-mpg.csv')
data.columns
```

```python
X = data[['weight','cylinders','model year']]
y = data['mpg']
```

```python
initial_model = sm.OLS(y, sm.add_constant(X)).fit()
initial_model.summary()
```

```python
X_metric = X.copy()
X_metric['weight'] = X_metric['weight'] * 0.453592
metric_model = sm.OLS(y, sm.add_constant(X_metric)).fit()
metric_model.summary()
```

```python
y_metric = y * 0.425144
model_metric = sm.OLS(y_metric, sm.add_constant(X_metric)).fit()
model_metric.summary()
```

```python
#shifting
X_shifted = X.copy()
X_shifted['model year'] = X_shifted['model year'] + 1900
shifted_model = sm.OLS(y, sm.add_constant(X_shifted)).fit()
shifted_model.summary()
```

Log Transformations

```python
# Log Transformations - A technique used to reshape data, especially when relationships between variables are non-linear
```

```python
plt.scatter(data['weight'], data['mpg'])
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.title('Raw Weight vs MPG')
plt.show()
```

```python
log_weight = np.log(data['weight'])
plt.scatter(log_weight, data['mpg'])
plt.xlabel('Log(Weight)')
plt.ylabel('MPG')
plt.title('Log(Weight) vs MPG')
plt.show()
```

```python
X_raw = sm.add_constant(data[["weight", "model year"]])
y = data['mpg']
raw_model = sm.OLS(y, X_raw).fit()
raw_model.summary()
```

```python
X_log = data[["weight", "model year"]].copy()
X_log['weight'] = np.log(X_log["weight"])
X_log = sm.add_constant(X_log)
log_model = sm.OLS(y, X_log).fit()
log_model.summary()
```

Polynomial Regression -  A method used to model the relationship between an input and output variable when the relationship is non-linear

e.g
Linear regression = y = B0 + B1x
Polynomial regression(Degree 2) = y = B0 + B1x + B2x^2
Polynomial regression(Degree 3) = y = B0 + B1x + B2x^2 + B3x^3


```python
yield_data = pd.read_csv('yield.csv', sep='\s+', index_col=0)
yield_data
```

```python
X = yield_data[['Temp']]
y = yield_data['Yield']
```

```python
plt.scatter(X, y)
plt.xlabel('Temp')
plt.ylabel('Yield')
plt.title('Temp vs Yield')
plt.show()
```

```python
linear_model = sm.OLS(y, sm.add_constant(X)).fit()
linear_model.summary()
```

```python
plt.scatter(X, y)
plt.plot(X, linear_model.predict(sm.add_constant(X)))
plt.title('Linear Regression Fit Line')
plt.show()
```

```python
X_quad = X.copy()
X_quad['Temp_squared'] = X_quad['Temp'] ** 2
```

```python
quad_model = sm.OLS(y, sm.add_constant(X_quad)).fit()
quad_model.summary()
```

```python
X_smooth = pd.DataFrame({'Temp':np.linspace(50,100,50)})
X_smooth['Temp_squared'] = X_smooth['Temp'] ** 2
```

```python
plt.scatter(X['Temp'], y, color='green')
plt.plot(X_smooth['Temp'], quad_model.predict(sm.add_constant(X_smooth)), color="blue")
plt.title('Quadratic Regression Fit Line')
plt.show()
```

Interaction Terms - In a regression model it allows the effect of one feature to depend on the value of another feature.

```python
data = pd.read_csv('auto-mpg.csv')
data.head()
```

```python
data['weight_centered'] = data['weight'] - data['weight'].mean()
data['model year']  = data['model year'] + 1900
```

```python
data.head()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
scatter = ax.scatter(data['weight_centered'],data['mpg'] ,c = data['cylinders'], cmap='plasma')
legend = ax.legend(*scatter.legend_elements(), title = 'Cylinders')
ax.add_artist(legend)
ax.set_xlabel("Centered Weight")
ax.set_ylabel("MPG")
ax.set_title("MPG VS Weight")
plt.tight_layout()
plt.show()
```

```python
X_no_interactions = data[['weight_centered','cylinders']].copy()
X_no_interactions = pd.get_dummies(X_no_interactions, columns=['cylinders'], drop_first=True)
X_no_interactions = sm.add_constant(X_no_interactions)

y = data['mpg']
```

```python
no_interaction_model = sm.OLS(y, X_no_interactions).fit()
no_interaction_model.summary()
```

```python
X_interaction = X_no_interactions.copy()
X_interaction['weight * cylinders_4'] = X_interaction['weight_centered'] * X_interaction['cylinders_4']
```

```python
interaction_model = sm.OLS(y, X_interaction).fit()
interaction_model.summary()
```

```python
X_interaction_8 = X_no_interactions.copy()
X_interaction_8['weight * cylinders_8'] = X_interaction_8['weight_centered'] * X_interaction_8['cylinders_8']
interaction_model_8 = sm.OLS(y, X_interaction_8).fit()
interaction_model_8.summary()
```
